While evaluating columns, it came to attention that certain columns gives us percentage values (i.e vitamin a, vitamin b and calcium) which makes it difficult to apply operations on them, so we convert the respective columns into integers by using astype() function and also remove the '%' symbol using str.rstrip().

```python
starbucks_df['vitamin_a'] = starbucks_df['vitamin_a'].str.rstrip('%').astype('int')
```

```python
starbucks_df['vitamin_c'] = starbucks_df['vitamin_c'].str.rstrip('%').astype('int')
```

```python
starbucks_df['calcium'] = starbucks_df['calcium'].str.rstrip('%').astype('int')
```

We can also add the '%' symbol to the column heading wherever necessary which tells us the unit certain constituents are measured in.

```python
starbucks_df=starbucks_df.rename(columns={'vitamin_a':'vitamin_a_%','vitamin_c':'vitamin_c_%','calcium':'calcium_%'})
```

Q4. Let us find the average Calories and Total Carbohydrates present in all Beverage Categories and visualize it for better understanding.

```python
starbucks_Beverage_category_df= starbucks_df.groupby('Beverage_category')[['Calories','total_carbohydrates_g']].mean()
```

```python
starbucks_Beverage_category_df
```

```python
sns.set_style("darkgrid")
```

```python
plt.title("Beverage Vs Calories Vs Total Carbohydrates")
plt.hist(starbucks_Beverage_category_df.Calories,alpha=0.4,bins=np.arange(2, 300, 20));
plt.hist(starbucks_Beverage_category_df.total_carbohydrates_g,alpha = 0.4,bins=np.arange(2, 300, 20));
```

This plot helps us visualize the average distribution of calories and total carbohydrates in the starbucks dataset.

We can also show to graph between beverage category and calories (alone) using a bar plot.

```python
plt.figure(figsize=(12,12));
plt.title("Beverage Vs Calories")
sns.barplot(x= 'Calories',y = 'Beverage_category',hue ='Beverage_category',data =starbucks_Beverage_category_df );
plt.legend(['Classic Espresso Drinks', 'Coffee', 'Frappuccino® Blended Coffee','Frappuccino® Blended Crème','Frappuccino® Light Blended Coffee','Shaken Iced Beverages','Signature Espresso Drinks','Smoothies','Tazo® Tea Drinks']);
```

A bar plot shows the relationship between a numerical variable and a categorical variable. A histogram, on the other hand, is used to represent the distribution of a continuous numerical variable.
